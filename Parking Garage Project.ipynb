# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
class ParkingGarage():
    def __init__(self):
        self.tickets = [number for number in range(0,3)]
        self.parkingSpace = [number for number in range(0,3)]
        self.currentTicket = {
            'ticket_numbers' : [None],
            'paid' : False
        }
        
    
    def takeTicket(self):
        ticket_number = self.tickets.pop()
        parkingspace_number = self.parkingSpace.pop()
        self.currentTicket['ticket_numbers'] = ticket_number
        if len(self.tickets) == 0 :
            print("Sorry, ther are no available spaces in the parking garage now. ")
        else:
            print(f"Your ticket number is {ticket_number} and parking space is {parkingspace_number}. ")
            print(f"Available spot: {len(self.parkingSpace) - 1} ")


    def payForParking(self):
        ticket_num = int(input("Please enter your ticket number. "))
        if ticket_num in self.currentTicket.values():
            amount = input("Please enter the amount you need to pay. ")
            if len(amount) > 1 :
                print("Your payment is confirmed. Please leave the parking garage in 15 minutes. ")
                self.currentTicket['paid'] = True 
                print("Thank you for choosing us!")
            else:
                print("Kindly settle your payment.")
        else:
            print("Invalid number! Please check and try again.")
    

    def leaveGarage(self):
        ticket_num = int(input("Please enter your ticket number. "))
        if ticket_num in self.currentTicket.values():
                if False == self.currentTicket['paid']:
                    print("Please settle your payment before departing. ")
                else:
                    print("Thank you, have a nice day! ")
                    
                    self.tickets.append(ticket_num)
                    self.parkingSpace.append(ticket_num)

                    print(f"Total ticket is updated: {len(self.tickets)}")
                    print(f"Total parking space is updated: {len(self.parkingSpace)}")
                    self.currentTicket['ticket_numbers'] = None
                    self.currentTicket['paid'] = False
        else:
            print("Invalid number! Please check and try again.")
        


parking = ParkingGarage()

def run():
    
    running = True

    while running:
        selection = input("Please select the service below: Park/ Pay/ Leave/ Quit. ")
        if selection.lower() == 'quit':
            break 

        elif selection.lower() == 'park':
            parking.takeTicket()

        elif selection.lower() == 'pay':
            parking.payForParking()
            
        elif selection.lower() == 'leave':
            parking.leaveGarage()
        
        else:
            print("This is not a invalid option. You can only enter Park/ Pay/ Leave/ Quit.")
           
run()
```
